```python
import pandas as pd
import plotly.graph_objects as go

# Assuming your DataFrame is named 'df'
# Reset the index to create a numeric index
df = df.reset_index(drop=True)

# Create a hover text string that includes date and prices for each point
hover_texts = df['date'].dt.strftime('%Y-%m-%d %H:%M') + '<br>Open: ' + df['open'].astype(str) + '<br>High: ' + df['high'].astype(str) + '<br>Low: ' + df['low'].astype(str) + '<br>Close: ' + df['close'].astype(str)

# Create an interactive candlestick chart using Plotly with custom hover text
fig = go.Figure(data=[go.Candlestick(x=df.index,
                                     open=df['open'],
                                     high=df['high'],
                                     low=df['low'],
                                     close=df['close'],
                                     hovertext=hover_texts,
                                     hoverinfo='text')])  # Use 'text' for custom hovertext


# Find the index positions of the middle of each trading day
middle_indices = df.groupby(df['date'].dt.date).apply(lambda x: x.index[len(x)//2]).values

# Create custom x-axis labels based on the middle index positions
x_labels = df.loc[middle_indices, 'date'].dt.strftime('%m-%d')

# Correctly find the numeric index positions of the first and last entries for each trading day
first_indices = df.groupby(df['date'].dt.date).apply(lambda x: x.index[0]).values
last_indices = df.groupby(df['date'].dt.date).apply(lambda x: x.index[-1]).values


# Create vertical lines for the start and end of each trading day
shapes = [dict(type='line',
               x0=i, y0=df['low'].min(), x1=i, y1=df['high'].max(),
               xref='x', yref='y',
               line=dict(color='green', width=1, dash='dash'))
          for i in first_indices]

# Customize the layout
fig.update_layout(
    title='Historical Prices',
    yaxis_title='Price',
    xaxis=dict(
        tickmode='array',
        #tickvals=middle_indices,
        #ticktext=x_labels,
        tickangle=-45
    ),
    xaxis_rangeslider_visible=False,
    #shapes=shapes
)

# Display the interactive plot
fig.show()
```

```python
import statsmodels.api as sm
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Assuming your DataFrame is named 'df' and contains a 'close' column
df['index'] = df.index.astype(int)

# cos_index
df['cos_index'] = np.cos(df.index)
df['sin_index'] = np.sin(df.index)
df['prev_close'] = df['close'].shift(1)

# Drop any missing values from the DataFrame
data = df.dropna()

# Prepare the data for linear regression
X = data[['index', 'cos_index', 'sin_index', 'volume']]
X = sm.add_constant(X)
y = data['close']

# Fit the linear regression model
model = sm.OLS(y, X).fit()
print(model.summary())

# Make predictions using the model
predictions = model.predict(X)

# Create a new DataFrame to store the predictions
pred_df = pd.DataFrame({'close': data['close'], 'predictions': predictions}, index=data.index)

# Create the line plot on the closing prices
fig = go.Figure(data=go.Scatter(x=df.index, y=df['close'], mode='lines', name='Closing Prices'))

# Add the predicted values to the plot
fig.add_trace(go.Scatter(x=pred_df.index, y=pred_df['predictions'], mode='lines', name='Predicted Prices'))

# Customize the layout
fig.update_layout(title='Closing Prices with Linear Regression',
                  xaxis_title='Time',
                  yaxis_title='Price')

# Display the interactive plot
fig.show()
```

```python
import pytz

# Assuming your SPY DataFrame is named 'spy_df'
# and the date column is named 'date'

# Convert the 'date' column to datetime if it's not already
spy_df['date'] = pd.to_datetime(spy_df['date'])

# Convert the time zone from US/Eastern to US/Central
central = pytz.timezone('US/Central')
spy_df['date'] = spy_df['date'].dt.tz_convert(central)

# Print the updated 'spy_df' DataFrame
spy_df.tail(50)
```

```python
start_date = '2022-12-27'
spy_df = spy_df[spy_df['date'] >= start_date]

# Reset the index
spy_df = spy_df.reset_index(drop=True)
spy_df.head(5)
```

```python
import pandas as pd
import plotly.graph_objects as go
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Assuming your DataFrames are named 'spy_df' and 'es_df'
# and the closing price columns are named 'close'

# Merge the DataFrames based on the 'date' column
merged_df = pd.merge(df, spy_df, on='date', suffixes=('_ES', '_SPY'), how='left')
merged_df.head(10)
```

```python
# Drop all colums except date, close_ES, close_SPY, and index
merged_df = merged_df[['date', 'close_ES', 'close_SPY', 'index']]
merged_df.head(10)
```

```python
# fill nan with the previous value
merged_df = merged_df.fillna(method='ffill')
merged_df.head(10)
```

```python
# Standard normalization
scaler_std = StandardScaler()
merged_df['close_ES_std'] = scaler_std.fit_transform(merged_df[['close_ES']])
merged_df['close_SPY_std'] = scaler_std.fit_transform(merged_df[['close_SPY']])
merged_df.head(10)
```

```python
# Min-max normalization
scaler_mm = MinMaxScaler()
merged_df['close_ES_mm'] = scaler_mm.fit_transform(merged_df[['close_ES']])
merged_df['close_SPY_mm'] = scaler_mm.fit_transform(merged_df[['close_SPY']])
merged_df.head(10)
```

```python
# Create the first graph with standard normalization
fig_std = go.Figure()
fig_std.add_trace(go.Scatter(x=merged_df['date'], y=merged_df['close_SPY_std'], name='SPY'))
fig_std.add_trace(go.Scatter(x=merged_df['date'], y=merged_df['close_ES_std'], name='ES'))
fig_std.update_layout(title='SPY vs ES Closing Prices (Standard Normalization)',
                      xaxis_title='Date',
                      yaxis_title='Normalized Price')
```

```python
# Create the second graph with min-max normalization
fig_mm = go.Figure()
fig_mm.add_trace(go.Scatter(x=merged_df['date'], y=merged_df['close_SPY_mm'], name='SPY'))
fig_mm.add_trace(go.Scatter(x=merged_df['date'], y=merged_df['close_ES_mm'], name='ES'))
fig_mm.update_layout(title='SPY vs ES Closing Prices (Min-Max Normalization)',
                     xaxis_title='Date',
                     yaxis_title='Normalized Price')
```
