```python
import pandas as pd

original_data_path = r"c:\Users\kmfoc\Downloads\nasdaq_external_data.csv"
new_data_path = r"C:\Users\kmfoc\My Drive\BU Files\BU Masters\Work\Personal\FNSPID_Financial_News_Dataset\data_scraper\headline scraper\combined_headlines_with_text.csv"
```

```python
original_df = pd.read_csv(original_data_path)
original_df.head()
```

```python
# drop unnamed
original_df = original_df.drop(columns=["Unnamed: 0"])
original_df.head()
```

```python
new_df = pd.read_csv(new_data_path)
new_df.head()
```

# simple combine

Only keep date, symbol, and text

```python
# original
# Only choose a subset of columns
original_df = original_df[["Date", "Stock_symbol", "Article"]]
original_df.head()
```

```python
# new
new_df = new_df[["Date", "Symbol", "Article_Text"]]
new_df.head()
```

```python
# rename columns
new_df = new_df.rename(columns={"Article_Text": "Text"})
original_df = original_df.rename(columns={"Stock_symbol": "Symbol", "Article": "Text"})

# Update the date format to (mm/dd/yyyy)
original_df["Date"] = pd.to_datetime(original_df["Date"]).dt.strftime('%m/%d/%Y')
new_df["Date"] = pd.to_datetime(new_df["Date"]).dt.strftime('%m/%d/%Y')

# Merge the data
merged_df = pd.concat([original_df, new_df])
merged_df.head()
```

```python
save_path = "../data/news_data_v2.csv"

# save
merged_df.to_csv(save_path, index=False)
```
