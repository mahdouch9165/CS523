```python
import pandas as pd

path = "../data/es_days_03_13_2024.csv"

df= pd.read_csv(path, encoding="utf-8", index_col=0)
df.head()
```

```python
df.tail()
```

```python
# news_df
news_path = "../data/news_data_v2.csv"

news_df = pd.read_csv(news_path, encoding="utf-8")
news_df.head()
```

```python
# group by date, and calculate the mean for negative neutral and positive
news_df['Date'] = pd.to_datetime(news_df['Date'])
grouped_news_df = news_df.drop(columns=["Symbol", "Text", "input_ids", "attention_mask"]).groupby('Date').mean()
# reset index
grouped_news_df.reset_index(inplace=True)
grouped_news_df.head()
```

```python
# describe the Date column
grouped_news_df.describe()
# 03/14 is the latest date
```

```python
# get the latest date in df
print(df['date'].max(), df['date'].min())
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
# filter df to dates greater than 2022-12-27
df = df[df['date'] >= "2022-12-27"]
df.head()
```

```python
# merge the two dataframes
merged_df = pd.merge(df, grouped_news_df, how="left", left_on="date", right_on="Date")
merged_df.head()
```

```python
merged_df.tail()
```

```python
# Make sure there are no discrepancies between date and Date
merged_df[merged_df['date'] != merged_df['Date']]
```

```python
# drop the Date column
merged_df.drop(columns=["Date"], inplace=True)
```
